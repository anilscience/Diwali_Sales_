# Exploratory Data Analysis (EDA)

## Gender

```python
xy = sns.countplot(x = "Gender", data = ds)

for bars in xy.containers:
    xy.bar_label(bars,color='r')
```

```python
sales_gen = ds.groupby(["Gender"], as_index = False) ["Amount"].sum().sort_values(by = "Amount", ascending = False)

xy = sns.barplot(x = "Gender", y = "Amount", data = sales_gen)
```

#### From the above-mentioned graphs, it is evident that the majority of the purchasers are female and that the purchasing power of females surpasses that of males.

# States

```python
ds.columns
```

```python
# Checking total number of orders from top 12 states 

sales_state = ds.groupby(["State"], as_index = False) ["Orders"].sum().sort_values(by = "Orders", ascending = False).head(12)
sns.set(rc={"figure.figsize":(19,5)})
xy= sns.barplot(x = "State", y = "Orders", data =sales_state)
```

```python
# Checking top 12 states based on Amount spent
sales_state = ds.groupby(["State"], as_index = False) ["Amount"].sum().sort_values(by = "Amount", ascending = False).head(12)

sns.set(rc={"figure.figsize":(19,5)})
xy = sns.barplot(x = "State", y = "Amount", data = sales_state)
```

##### From the above graphs we can see that the states of Uttar Pradesh, Maharashtra, and Karnataka contribute significantly to the majority of the orders and total sales/amount.


# Marital Status


```python
xy = sns.countplot(x ='Marital_Status', data = ds,color='y')

sns.set(rc={"figure.figsize":(8,5)})
for bars in xy.containers:
    xy.bar_label(bars,color='g')
```

```python
sales_state = ds.groupby(["Marital_Status", "Gender"], as_index = False) ["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.set(rc={"figure.figsize":(8,5)})
xy = sns.barplot(x = "Marital_Status", y = "Amount", data = sales_state, hue = "Gender")
```

##### From the above graphs, it is evident that the majority of the buyers are married women, and they possess a high purchasing power.

# Occupation

```python
xy = sns.countplot(x = "Occupation", data = ds)

sns.set(rc={"figure.figsize":(20,5)})
for bars in xy.containers:
    xy.bar_label(bars,color='r')
```

```python
sales_state = ds.groupby(["Occupation"], as_index = False) ["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.set(rc={"figure.figsize":(20,5)})
xy = sns.barplot(x = "Occupation", y = "Amount", data = sales_state)
```
